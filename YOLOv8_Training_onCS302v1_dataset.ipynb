# Model Training #1

We will start with the model with pretrained weights. Ultralytics recommeded this model for fine tuning.

# Model Training #3

Trying another approach by combining the previous two in order to find out the best possible architecture for a custom-trained model.

```python
model = YOLO('yolov8n.yaml').load('yolov8n.pt')
```

```python
model.train(data="/content/drive/MyDrive/CS302.v1i.yolov8/data.yaml", epochs=100)
```

By looking at the visualizations of results saved in the runs folder, the model did not seem to converge, so I decided to continue training for 50 more epochs.

```python

model = YOLO('yolov8n.yaml').load('/content/runs/detect/train4/weights/best.pt')

# Continue training for additional epochs
model.train(data="/content/drive/MyDrive/CS302.v1i.yolov8/data.yaml", epochs=50)
```
